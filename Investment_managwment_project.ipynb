```python
import yfinance as yf
import pandas as pd
from scipy.stats import zscore

# Universe stocks

# Get book/price data
bp_data = pd.DataFrame()

for ticker in tickers:
    tkr = yf.Ticker(ticker)
    book_value = tkr.info['bookValue']
    price = tkr.info['previousClose']
    bp = book_value / price
    bp_data = bp_data.append({'ticker':ticker,'bp':bp},ignore_index=True)

# Calculate z-scores of BP
bp_zscores = zscore(bp_data['bp'])
bp_data = bp_data.assign(zscore=bp_zscores)

# Top/bottom decile
num_stocks = len(tickers)
top_decile = bp_data.sort_values('zscore', ascending=False).iloc[:5]
bot_decile = bp_data.sort_values('zscore').iloc[:5]

# Over/under weights
top_weight = 0.6 / len(top_decile)
bot_weight = 0.1 / len(bot_decile)

neutral_weight = (1 - top_weight*len(top_decile) - bot_weight*len(bot_decile)) / (num_stocks - len(top_decile) - len(bot_decile))

# Construct full weights
top_weights = [top_weight]*len(top_decile)
bot_weights = [bot_weight]*len(bot_decile)
neutral_weights = [neutral_weight] * (num_stocks - len(top_decile) - len(bot_decile))

all_weights = dict(zip(top_decile['ticker'], top_weights))
all_weights.update(dict(zip(bot_decile['ticker'], bot_weights)))
all_weights.update(dict(zip(list(set(tickers) - set(top_decile['ticker']) - set(bot_decile['ticker'])), neutral_weights)))

print(all_weights)
```

```python
import yfinance as yf
import pandas as pd

# Universe of tickers


# Get returns data
ticker_returns = {t: yf.Ticker(t).history(period="max")[['Close']].pct_change().dropna() for t in tickers if len(yf.Ticker(t).history(period="3y"))>0}

# Benchmark
benchmark = yf.Ticker("^GSPC").history(period="max")[['Close']].pct_change().dropna()

# Calculate relative momentum
rel_momentum = {t:ret.sub(benchmark,axis=0)['Close'].mean()*252 for t,ret in ticker_returns.items()}

# Calculate absolute momentum
abs_momentum = {t:ret['Close'].dropna().iloc[-252:].mean()*252 for t,ret in ticker_returns.items() if len(ret) > 252}

# Apply dual momentum strategy
selected = {t:rel_mom for t,rel_mom in rel_momentum.items() if t in abs_momentum and abs_momentum[t]>0}
tot_mom = sum(selected.values())
weights = {t:m/tot_mom for t,m in selected.items()}

# Create dataframe with all tickers
df = pd.DataFrame(columns=['Ticker', 'Weight'])
for t in tickers:
    df = df.append({'Ticker':t}, ignore_index=True)

# Populate weights for selected tickers
for t in selected.keys():
    df.loc[df['Ticker']==t, 'Weight'] = weights[t]

# Fill NA with zero
df['Weight'].fillna(0, inplace=True)

print(df)
```

```python
import yfinance as yf
import numpy as np


# Get valuation ratios from yfinance
tpe, fpe = [], []
for ticker in tickers:
    ticker_info = yf.Ticker(ticker).info

    try:
        tpe.append(ticker_info['trailingPE'])
    except KeyError:
        tpe.append(None)

    try:
        fpe.append(ticker_info['forwardPE'])
    except KeyError:
        fpe.append(None)

# Calculate ratios
ratios = [tpe[i]/fpe[i] if fpe[i] else None for i in range(len(tickers))]

# Target allocations
target_alloc = 8000
amounts = [ratios[i] * target_alloc if ratios[i] else target_alloc for i in range(len(tickers))]
amounts = [min(amt,12000) if amt else amt for amt in amounts]

# Weights
total_alloc = np.sum(amounts)
weights = [amount / total_alloc for amount in amounts]

value_weights = dict(zip(tickers, weights))
print(value_weights)
```
